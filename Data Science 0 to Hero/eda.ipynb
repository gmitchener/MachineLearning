```python
df = df.drop(['ID', 'DOB', 'CollegeID', '12graduation', 'GraduationYear', '10board', '12board', 'CollegeState', 
'CollegeCityID', 'CollegeCityTier'], axis=1)
df.head()
```

```python
df.duplicated().sum()
```

```python
df.Specialization.value_counts()
```

```python
specialization = df.Specialization.value_counts()
lessthan10 = specialization[specialization<=10]
lessthan10
```

```python
correctSpecialization_map = dict(zip(lessthan10, 'Other'))
```

```python
df.Specialization = df['Specialization'].apply(lambda x: 'Other' if x in lessthan10 else x)
```

```python
df.describe()
```

```python
def scatter_plot(column):
    plt.scatter(df.index, df[column])
    plt.ylabel(column)
    plt.show()
```

```python
scatter_plot('collegeGPA')
```

```python
df = df[(df.collegeGPA > 40)]
```

```python
scatter_plot('Domain')
scatter_plot('ComputerProgramming')
scatter_plot('ElectronicsAndSemicon')
```

```python
cols_with_error = ['Domain', 'ComputerProgramming', 'ElectronicsAndSemicon', 'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg']

for col in cols_with_error:
    df[col] = df[col].apply(lambda x: np.nan if x == -1 else x)

for col in cols_with_error:
    df[col] = df[col].fillna(df[col].mean())
```

```python
scatter_plot('ComputerProgramming')
```

### Data Visualisation

1. Univariate Analysis: understand the distribution of values for a single variable

```python
sns.distplot(df.charges)
plt.show()

sns.boxplot(df.charges)
plt.show()
```

```python
Q1 = df.charges.quantile(0.25)
Q3 = df.charges.quantile(0.75)
IQR = Q3-Q1 
print(f'Interquartile range for "charges" column: {IQR}')

df.charges = df['charges'].apply(lambda x: np.nan if (x < Q1-1.5*IQR) or (x > Q3+1.5*IQR) else x)
df.dropna(inplace=True)

sns.boxplot(df.charges)
plt.show()
```

```python
sns.histplot(df.age)
plt.show()
```

```python
sns.histplot(df.bmi)
plt.show()
```

```python
def categorical_bar_plot(column):
    plt.bar(dict(df[column].value_counts()).keys(), dict(df[column].value_counts()).values())
    plt.xlabel(column)
    plt.show()
```

```python
categorical_bar_plot('sex')
categorical_bar_plot('children')
categorical_bar_plot('smoker')
categorical_bar_plot('region')
```
